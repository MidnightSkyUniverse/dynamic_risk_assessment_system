## F1 scores

```python
f1_scores = db_select("SELECT f1_score from f1;")
f1list = [f[0] for f in f1_scores]

plt.figure(figsize=(12,6))
plt.title("F1 scores")
plt.plot(f1list)
```

## Mean, median and std for lastmonth_activity

```python
numeric_cols = ["lastmonth_activity", "lastyear_activity", "number_of_employees"]
mean = db_select(f"SELECT mean from feature_stats where feature='{numeric_cols[0]}';")
median = db_select(f"SELECT median from feature_stats where feature='{numeric_cols[0]}';")
std = db_select(f"SELECT std from feature_stats where feature='{numeric_cols[0]}';")
mean_ = [f[0] for f in mean]
median_ = [f[0] for f in median]
std_ = [f[0] for f in std]

plt.figure(figsize=(12,6))
plt.title(f"Column: {numeric_cols[0]}")
plt.plot(mean_)
plt.plot(median_)
plt.plot(std_)
plt.legend(['mean','median','std'])
```
